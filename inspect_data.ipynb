```python
from pydantic import BaseModel

class Item(Info):
    tradable: bool
```

```python
from relevants.cs2_terms import rarity
rarity_l = rarity.strip().splitlines()
rarity_rank = {r: i for i, r in enumerate(rarity_l)}
def add_tradable_state(mapping):
    new_mapping = {}
    for container, weapons in mapping.items():
        ranks = [rarity_rank[weapon.data.goods_info.rarity_localized_name] for weapon in weapons]
        highest = max(ranks)
        items = []
        for weapon in weapons:
            if rarity_rank[weapon.data.goods_info.rarity_localized_name] == highest:
                tradable = False
                items.append(Item(**weapon.model_dump(), tradable=tradable))
            else:
                tradable = True
                items.append(Item(**weapon.model_dump(), tradable=tradable))
        new_mapping[container] = items
    return new_mapping
mappings = {
    'stattrack': add_tradable_state(stat_mapping),
    'normal': add_tradable_state(normal_mapping)
}
```

#### Build a dictionary which every container has a hierarchy of weapons

```python
from operator import attrgetter
get_rarity = attrgetter('data.goods_info.rarity_localized_name')
def build_hierarchy(mapping: dict[str, list[Item]]):
    hierarchy = {}
    for container, items in mapping.items():
        by_rarity = {}
        for item in items:
            rarity = get_rarity(item)
            if rarity not in by_rarity:
                by_rarity[rarity] = []
            by_rarity[rarity].append(item)
        hierarchy[container] = by_rarity
    return hierarchy
hierarchies = {
    'stattrack': build_hierarchy(mappings['stattrack']),
    'normal': build_hierarchy(mappings['normal'])
}
```

```python
from collections import Counter
from recipe_optimize.ga_fitness import container_possibilities, weapon_float
from recipe_optimize.utilities import Weapon, return_hash_name, wear_rank_p
from relevants.cs2_terms import *

def modify_value_getter(value_getters):
    def modified_getter(item: Item):
        for getter in value_getters:
            value = getter(item)
            if value:
                return value
        return 0
    return modified_getter

def modify_float_getter(float_getter):
    # return average float to represent the float value
    def modified_getter(item: Item):
        if all(value_t:=float_getter(item)):
            return sum(value_t)/2
        wear_rank = wear_rank_p.search(item.data.goods_info.market_hash_name).group(1)
        wear_range = wear_value.get(wear_rank)
        min_, max_ = list(map(float, wear_range.split('-')))
        return (min_ + max_) / 2
    return modified_getter

def fitness(weapons: list[Item], rarity_rank: int, container_identity_info: dict, container_price_info: dict):

    # get container statistics
    container_list = list()
    for weapon in weapons:
        container_list.append(weapon.data.container[0].name)
    c = Counter(container_list)
    container_names = list(c.keys())
    next_rarity = rank_to_rarity[rarity_rank + 1]
    identity_info: dict[str, list[Weapon]] = {k: container_identity_info[k][next_rarity] for k in container_names}
    price_info: dict[str, list[Item]] = {k: container_price_info[k][next_rarity] for k in container_names}

    # get correspond possiblity of outcomes
    next_rarity_num = {}
    for container in container_names:
        next_rarity_num[container] = len(identity_info[container])
    containers = {container: (c[container], next_rarity_num[container]) for container in c}
    possibility_of_container = container_possibilities(containers)

    # get expected value and profit percentage
    float_getter = modify_float_getter(attrgetter('data.goods_info.min_float', 'data.goods_info.max_float')) # we use min float as the representative float value
    value_getter = modify_value_getter((attrgetter('data.goods_info.yyyp_sell_price'), attrgetter('data.goods_info.buff_sell_price')))# we use 悠悠有品 as the representative price, fallback to buff price
    name_getter = attrgetter('data.goods_info.market_hash_name')
    floats = list(map(float_getter, weapons))
    expected_value = 0
    for container in container_names:
        for weapon in identity_info[container]:
            float_ = weapon_float(floats, weapon.min_float, weapon.max_float)
            hash_name = return_hash_name(weapon.weapon_name, float_)
            value = None
            for item in price_info[container]:
                if name_getter(item) == hash_name:
                    value = value_getter(item)
                    break
            if not value:
                return -1
            expected_value += value * possibility_of_container[container]
    cost = 0
    for weapon in weapons:
        price = value_getter(weapon)
        if not price:
            return -1
        cost += price
    income = expected_value - cost
    return income
```

```python
def generate_combinations(total_balls, num_bins):
    """
    Generate all combinations of distributing identical balls into different bins.
    Each bin must have at least 1 ball.
    
    Args:
        total_balls: Total number of identical balls
        num_bins: Number of different bins
    
    Yields:
        Tuple representing the number of balls in each bin
    """
    if num_bins > total_balls:
        return  # Impossible to put at least 1 ball in each bin
    
    if num_bins == 1:
        yield (total_balls,)
        return
    
    # Recursively distribute balls
    for first_bin in range(1, total_balls - num_bins + 2):
        remaining_balls = total_balls - first_bin
        remaining_bins = num_bins - 1
        
        # Generate all combinations for the remaining bins
        for rest in generate_combinations(remaining_balls, remaining_bins):
            yield (first_bin,) + rest
```

#### Method 2 (case combination + local optimized)
We only considering the combination of 1/2/3 cases

```python
keys = list(hierarchies['normal'].keys())
len(keys)
```

```python
import itertools
combinations = []
for number in [1,2,3]:
    for comb in itertools.combinations(keys, number):
        combinations.append(comb)
```

```python
from tqdm import tqdm
def validate(containers:tuple, weapons_price_info: dict, rarity: str):
    flag = None
    for container in containers:
        if weapons_price_info[container].get(rarity, None) is None:
            flag = False
            break
        elif weapons_price_info[container][rarity][0].tradable:
            flag = True
        else:
            flag = False
            break
    return flag

def get_best(container_comb, weapon_price_info, rarity):
    if not validate(container_comb, weapon_price_info, rarity):
        return
    value, list_ = get_best_from_combination(container_comb, weapon_price_info, rarity)
    return value, list_, container_comb

def get_best_from_combination(container_comb, weapons_price_info, rarity):
    def gen_list():
        weapon_candidates = {k: weapons_price_info[k][rarity] for k in container_comb}
        # Get all weapon combinations by taking one from each container in weapon_candidates
        weapon_combinations = itertools.product(*weapon_candidates.values())
        num_containers = len(container_comb)
        for weapon_combination in weapon_combinations:
            for distribution in generate_combinations(10, num_containers):
                weapon_list = []
                weapon_num = zip(weapon_combination, distribution)
                for weapon, number in weapon_num:
                    weapon_list.extend([weapon for _ in range(number)])
                yield weapon_list
    g = gen_list()
    max_value = float('-inf')
    for list_ in g:
        value = fitness(list_, rarity_rank[rarity], container_identity, hierarchies['normal'])
        if value > max_value:
            max_value = value
            best_list = list_
    return max_value, best_list

        
```

```python
rank = 5
weapons_price_info = hierarchies['normal']
rarity = rank_to_rarity[rank]

# from joblib import Parallel, delayed
# results = Parallel(n_jobs=4)(delayed(get_best)(item, weapons_price_info, rarity) for item in tqdm(combinations))
map_obj = map(get_best, combinations, itertools.repeat(weapons_price_info), itertools.repeat(rarity))
results = []
for result in tqdm(map_obj, total=len(combinations)):
    results.append(result)
 
```
